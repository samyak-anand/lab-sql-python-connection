```python
from sqlalchemy import text



def rental_month(engine, month,year):
    with engine.connect() as connection:
        txt= f''' SELECT * FROM rental WHERE EXTRACT(MONTH FROM rental_date) = {month} AND EXTRACT(YEAR FROM rental_date) = {year}'''
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
    
```

```python
def rental_count_by_month():
    # Establishes a connection to the database using the provided SQLAlchemy engine
    with engine.connect() as connection:
        
        # SQL query: 
        # - Extracts the month from each rental_date as a 2-digit string ('%m')
        # - Counts the number of rentals per month
        # - Groups results by the extracted month
        # - Orders the result by month (chronologically)
        txt = '''
            SELECT DATE_FORMAT(rental_date, '%m') AS month, COUNT(*) AS rental_count
            FROM rental
            GROUP BY month
            ORDER BY month;
        '''
        
        # Prepares the SQL query for execution using SQLAlchemy's `text()` method
        query = text(txt)
        
        # Executes the query on the connected database
        result = connection.execute(query)
        
        # Converts the result to a pandas DataFrame with appropriate column names
        return pd.DataFrame(result.all(), columns=['month', 'rental_count'])
```

```python
from sqlalchemy import text
import pandas as pd

def rental_count_month(month, year, engine):
    """
    Returns a DataFrame with number of rentals by customer for the given month and year.
    The column is dynamically named rentals_MM_YYYY.
    """
    month_str = f"{int(month):02d}"
    year_str = str(year)
    column_name = f"rentals_{month_str}_{year_str}"

    sql = f'''
        SELECT customer_id, COUNT(*) AS `{column_name}`
        FROM rental
        WHERE MONTH(rental_date) = :month AND YEAR(rental_date) = :year
        GROUP BY customer_id
        ORDER BY customer_id;
    '''

    with engine.connect() as connection:
        query = text(sql)
        result = connection.execute(query, {'month': int(month), 'year': int(year)})
        return pd.DataFrame(result.fetchall(), columns=['customer_id', column_name])
```

```python
def compare_rental():
    sql= ''' 
         SELECT
            m1.customer_id,
            m1.rentals AS rentals_05_2005,
            m2.rentals AS rentals_06_2005,
            (m2.rentals - m1.rentals) AS difference
        FROM
            (
                SELECT customer_id, COUNT(*) AS rentals
                FROM rental
                WHERE MONTH(rental_date) = 5 AND YEAR(rental_date) = 2005
                GROUP BY customer_id
            ) m1
        JOIN
            (
                SELECT customer_id, COUNT(*) AS rentals
                FROM rental
                WHERE MONTH(rental_date) = 6 AND YEAR(rental_date) = 2005
                GROUP BY customer_id
            ) m2
        ON m1.customer_id = m2.customer_id;
         '''
         
    with engine.connect() as connection:
        query = text(sql)
        result = connection.execute(query)
        return pd.DataFrame(result.fetchall())
```
